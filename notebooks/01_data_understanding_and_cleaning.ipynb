## Data Cleaning

The dataset has no missing values or duplicate records.
The only preprocessing required is converting date columns to the appropriate datetime format.

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=False)
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=False)
```

```python
df.info()
```

```python
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month_name()
df["Month Number"] = df["Order Date"].dt.month
df["Quarter"] = df["Order Date"].dt.quarter
```

```python
df.to_csv("../processed_data/Superstore_Cleaned.csv", index=False)
```
